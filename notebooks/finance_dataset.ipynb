```python
import yfinance as yf
import pandas as pd

def fetch_extended_data(ticker: str, country: str, period='2500d'):
    df = yf.download(ticker, period=period)
    
    if df.empty:
        print(f"‚ùå No data fetched for {country} ({ticker})")
        return None
    
    filename = f"{country}_{ticker.replace('^','')}_extended.csv"
    df.to_csv(filename)
    print(f"{len(df)} rows saved to {filename}")
    return df

# Fetch data for all 3 countries
if __name__ == "__main__":
    tickers = {
        "^NSEI": "India",
        "^ISEQ": "Ireland",
        "^GSPC": "USA"
    }

    dfs = {}
    for ticker, country in tickers.items():
        dfs[country] = fetch_extended_data(ticker, country)
```

### Moving Average

```python
def add_technical_indicators(df):
    df = df.copy()

    # Moving Averages
    df['MA_5'] = df['Close'].rolling(window=5).mean()
    df['MA_20'] = df['Close'].rolling(window=20).mean()
    
    # Rolling Volatility
    df['Vol_10'] = df['LogRet'].rolling(window=10).std()
    
    # RSI (Relative Strength Index)
    delta = df['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    
    rs = avg_gain / avg_loss
    df['RSI_14'] = 100 - (100 / (1 + rs))

    # MACD and Signal Line
    ema_12 = df['Close'].ewm(span=12, adjust=False).mean()
    ema_26 = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = ema_12 - ema_26
    df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

    return df
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Select relevant columns
features = ['Close', 'MA_5', 'MA_20', 'Vol_10', 'RSI_14', 'MACD', 'Signal']
IND_clean = IND.dropna()[features + ['Target']]  # ensure no NaNs

# Scale features
scaler_ind = MinMaxScaler()
X_scaled_ind = scaler_ind.fit_transform(IND_clean[features])
y_ind = IND_clean['Target'].values
```

```python
def create_windows(X, y, window_size=60):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        Xs.append(X[i:i+window_size])
        ys.append(y[i+window_size])
    return np.array(Xs), np.array(ys)

X_tensor_ind, y_tensor_ind = create_windows(X_scaled_ind, y_ind)
```

```python
def split_data(X, y, train_ratio=0.8):
    split = int(train_ratio * len(X))
    return X[:split], X[split:], y[:split], y[split:]

X_train_ind, X_test_ind, y_train_ind, y_test_ind = split_data(X_tensor_ind, y_tensor_ind)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
def build_cnn_lstm_model(input_shape):
    model = Sequential()
    
    # CNN block
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    
    # LSTM block
    model.add(LSTM(units=64, activation='tanh'))
    model.add(Dropout(0.2))
    
    # Output
    model.add(Dense(1))  # regression: next-day close
    
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
# Use your prepared tensors from earlier
# X_train_ind, X_test_ind, y_train_ind, y_test_ind

input_shape = X_train_ind.shape[1:]  # (60, 7)
model = build_cnn_lstm_model(input_shape)

# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train_ind, y_train_ind,
    validation_data=(X_test_ind, y_test_ind),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Predict on test set
y_pred_ind = model.predict(X_test_ind)

# Evaluation metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

rmse = np.sqrt(mean_squared_error(y_test_ind, y_pred_ind))
mae = mean_absolute_error(y_test_ind, y_pred_ind)

print(f"Test RMSE: {rmse:.2f}")
print(f"Test MAE: {mae:.2f}")
```
